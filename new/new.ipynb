```python
import pandas as pd
df = pd.read_csv('processed_data.csv')
# df.info()
df.columns
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import pickle
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = df.drop('Price ($)', axis=1)   # Dependant columns / Input columns
y = df['Price ($)']                # Resultant columns / Output column
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=0.3, shuffle=False) # 80:20, 70:30, 60:40
```

```python
model = KNeighborsRegressor(n_neighbors=2)
model.fit(X, y)

```

```python
y_pred = model.predict(X)
y_pred
```

```python
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R²: {r2:.2%}")

# Save the trained model
with open("knn_model.pkl", "wb") as f:
    pickle.dump(model, f)

print("Model saved successfully!")
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import pickle

# Load your dataset
data = pd.read_csv("processed_data.csv")

# Strip all column names
data.columns = data.columns.str.strip()

# Features and target
X = data[["Brand", "Model", "Storage", "RAM", "Screen Size (inches)",
          "Camera (MP)", "Battery Capacity (mAh)"]]
y = data["Price ($)"]

# Train the model
model = RandomForestRegressor()
model.fit(X, y)

# Save the model
with open("phone_price_model.pkl", "wb") as f:
    pickle.dump(model, f)
```
